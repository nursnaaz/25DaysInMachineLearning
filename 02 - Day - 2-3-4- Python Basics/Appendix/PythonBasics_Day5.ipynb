```python
def abs_value(num):
    """This function returns the absolute value of the entered number"""
    if(num>=0):
        return num
    else:
        return -num

print(abs_value(3))
print(abs_value(-3))
```

```python
#Argument/Parameter

def greet(name,msg):
    """This function greets to the person passed in as parameter"""
    print("Hello, ",name,msg)

#Function call
greet("xyz","Good Morning") 
greet("abc","Good Morning")

```

```python
def greet(name,msg = "Good morning"):
    """This function greets to the person passed in as parameter"""
    print("Hello, ",name,msg)

#Function call
greet("xyz","Good Night") 
greet("abc")

```

```python
#Arbitrary Arguments

def greet(*name):
    """This function greets to the person passed in as parameter"""
    for i in name:
        print("Hello, ",i)

#Function call
greet("xyz","noor","csac")
```

```python
def call(*name):
    j = []
    for i in name:  
        j.append(i*5)  
    return sum(j)
#Function call   
print(call(2,3,4))
```

```python
#Simple calculator

'''Program make a simple calculator'''
#This function adds the two numbers
def add(x,y):
    return x+y

#This function subtract the two numbers
def sub(x,y):
    return x-y

#This function multiplication the two numbers
def mul(x,y):
    return x*y

#This function division the two numbers
def div(x,y):
    return x/y

print("Select operation : ")
print("1. Add ")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
#Take input from the user
choice = input("Enter the choice (1/2/3/4): ")

num1 = int(input("Enter the first number "))
num2 = int(input("Enter the second number "))

if choice == "1":
    print(num1,"+",num2," = ",add(num1,num2))
    
elif choice == "2":
    print(num1,"-",num2," = ",sub(num1,num2))
    
elif choice == "3":
    print(num1,"*",num2," = ",mul(num1,num2))

elif choice == "4":
    print(num1,"/",num2," = ",div(num1,num2))

else:
    print("Invalid input")
```
